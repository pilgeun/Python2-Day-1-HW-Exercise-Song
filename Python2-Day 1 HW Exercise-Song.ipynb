## Exercise 1: 
Create a Player class with a name, a position, and a speed attributes. Set the initial value of speed to 1 and create any necessary methods.
Then, 
1. Create two Player objects and print information about both players.
2. Assign 8 to player 1's speed
3. print information about both players.

````python
class Player:
    """Player class has name, position, and speed attributes"""
    def __init__(self, name, position) :
        self.name = name
        self.position = position
        self.speed = 1
    
    def print_info(self) :
        print(f"{self.name} in Position {self.position} going at Speed {self.speed}")

player1 = Player("Mandy", 1)
player2 = Player("Billy", 2)

player1.speed = 8

player1.print_info()
player2.print_info()
````

## Exercise 2: 
Create a class called Rectangle with height and length attributes and has following methods: 
- perimeter() : returns a perimeter of a rectangle object
- area() : returns an area of a rectangle object

Then,
1. create two instances of Rectangle class
2. print each rectangle's perimeter and determine which rectangle has larger perimeter
3. print each rectangle's area and determine which rectangle has larger area

````python
class Rectangle:
    """Rectangle class has height and length attributes"""
    def __init__(self, height, length) :
        """Initialization method with height and width parameters"""
        self.height = height
        self.length = length
        
    def area(self) :
        return 2*self.height + 2*self.length

    def perimeter(self) :
        return self.height * self.length        

    
rec1 = Rectangle(10, 25)
rec2 = Rectangle(5, 50)

print(f"Perimeter of Rectangle 1 is {rec1.perimeter()}.")
print(f"Perimeter of Rectangle 2 is {rec2.perimeter()}.")
#Both rectangles have the same perimeter

print(f"Area of Rectangle 1 is {rec1.area()}.")
print(f"Area of Rectangle 2 is {rec2.area()}.")
#Rectangle 2 has the larger area
````

## Exercise 3: 
Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



````python
import math

class Triangle :
    """Triangle class has height and base attributes"""
    def __init__(self, height, base):
        """Initialization method with height and base parameters"""
        self.height = height
        self.base = base
        
    def perimeter(self):
        return self.height + self.base + math.sqrt((self.height**2)+(self.base**2))
        
    def area(self):
        return (self.height*self.base)/2

tri1 = Triangle(30, 40)
tri2 = Triangle(35, 35)

print(f"Perimeter of Triangle 1 is {tri1.perimeter():.2f}.")
print(f"Perimeter of Triangle 2 is {tri2.perimeter():.2f}.")
#Triangle 1 has the larger perimeter

print(f"Area of Triangle 1 is {tri1.area():.2f}.")
print(f"Area of Triangle 2 is {tri2.area():.2f}.")
#Triangle 2 has the larger area
````

## Exercise 4: 
Write a program that maintains a list of movies, and the user can list all the movies, add a movie to the list, or delete a movie from the list. All input/output are done at the program level, not at a class level.


For Python 1 Day 3 Challenge Exercise, you wrote a program that maintains a list of movies, and the user can list all the movies, add a movie to the list, or delete a movie from the list. Modify that program to use a <b>Movie object</b> to store name, year, and genre of a movie. Create any method you need to access these attributes but your method can not use <b>print</b> function. 

Additionally, create any function you need to make sure program neat. You may use <b>print</b> function in your function.


````python
class Movie :
    """Movie class has name, year and genre attributes"""
    def __init__(self, name, year, genre) :
        """Initialization method with name, year and genre parameters"""
        self.name = name
        self.year = year
        self.genre = genre
    

def add_movie(m):
    name = input("Name: ")
    year = input("Year: ")
    genre = input("Genre: ")
    new_Movie = Movie(name, year, genre)
    m.append(new_Movie)   
    print(new_Movie.name, "was added.") 

def delete_movie(m):
    list_number = int(input("Number: "))
    if list_number <= 0 or list_number > len(m):    # reject input if # in list does not exist
        print("Does not exist.")
    else:
        print(m[list_number-1].name, "was deleted.")
        m.pop(list_number-1)

def list_movie(m):

    print("#| Name \tYear \tGenre")
    for n in range(0, len(m)):   
        print(f"{n+1}. {m[n].name} \t{m[n].year} \t{m[n].genre}")
        
        
# creating initial list of movies
movie1 = Movie("Toy Story", 1995, "comedy")
movie2 = Movie("Lion King", 1994, "drama")
movie3 = Movie("Frozen", 2013, "fantasy")

movies = [movie1, movie2, movie3]
        
print("""
###### COMMAND MENU
```
list - list all movies
add - add a movie
del - delete a movie
exit - exit program
```
    """)

while True:
    command = input("###### Command: ")
    command = command.lower()
    print("```")
    
    if command == 'exit':
        break
    
    elif command == "list":
        list_movie(movies)
    
    elif command == "add":
        add_movie(movies)
                  
    elif command == "del": 
        delete_movie(movies)
    
    else:
        continue
        
    print("```\n")
    
    
````

````python
import random as rand # for assigning 'active' or 'not active'

class Employee :
    """Employee class has firstname, lastname, joined_date and active attributes"""
    def __init__(self, firstname, lastname, joined_date) :
        """Initialization method with firstname, lastname, joined_date and active parameters"""
        self.firstname = firstname
        self.lastname = lastname
        self.joined_date = joined_date
        self.active = True
    

def all_employees(muppets): 
    print("ALL EMPLOYEES")
    print("First\tLast\t\tStart")
    print("Name:\tName:\t\tdate:")
    for employee in muppets:
        print(f"{employee.firstname}\t{employee.lastname}\t\t{employee.joined_date}")
    print()
        

def active_employees(muppets): 
    print("CURRENTLY ACTIVE")
    print("First\tLast\t\tStart")
    print("Name:\tName:\t\tdate:")
    for employee in muppets:
        if employee.active == True:
            print(f"{employee.firstname}\t{employee.lastname}\t\t{employee.joined_date}")
    print()
           
        
data = [{"first":"elmo", "last":"monster", "year": 2000},
        {"first":"count", "last":"count", "year": 1819},
        {"first":"big", "last":"bird", "year": 2011},
        {"first":"cookie", "last":"monster", "year": 2002},
        {"first":"bert", "last": "person", "year": 2013},
        {"first":"ernie", "last": "person", "year": 2013},
        {"first":"oscar", "last": "grouch", "year": 1999},
        {"first":"zoe", "last": "cutie", "year": 2015},
        {"first":"abby", "last": "cadabby", "year": 2016},
        {"first":"prairie", "last": "dawn", "year": 2020}
   ]

employees = []

for muppet in data: # adds employee information objects
    new_employee = Employee(muppet["first"], muppet["last"], muppet["year"])
    employees.append(new_employee)

for employee in employees: # randomly selects employees and changes active status to 'False'
    is_active = rand.randint(0,1)
    if is_active == 0:
        employee.active = False


all_employees(employees) 
active_employees(employees)
````
